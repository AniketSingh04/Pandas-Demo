```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.random.rand(10,5), index = np.arange(10))
```

```python
df
```

**loc function**

```python
row = 1
column = 3
df.loc[row][column]
```

```python
df.loc[row,column]
```

```python
df.loc[[1,2],[3,4]]
```

```python
df.loc[:,3]
```

```python
df.loc[3,:]
```

```python
#Updating DataFrame using loc() function
df.loc[1,2] = 15
df
```

```python
df.loc[1:5,1:3] = 67
df
```

```python
df.iloc[1:5,2:4]
```

```python
df.loc[1:5]
```
